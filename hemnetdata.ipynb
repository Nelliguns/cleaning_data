```python
def find_int(x):
    digits = ""
    try:
        for char in x:
            if char.isdigit():
                digits.join(char)
        return digits
    except Exception as e:
        print(e)

```

```python
for char in spec_chars:
    df_copy['property_size'] = df_copy['property_size'].str.replace(char, ' ')
```
